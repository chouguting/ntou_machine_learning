```python
import numpy as np
from math import factorial as fac

def binomialDistribution_greater_than_e(N, e):
    p=0.1
    SUM=0

    print("(%d, %d):"%(N, e), end='')
    for i in np.arange(e, N+1):
        SUM+=(fac(N)/(fac(i)*fac(N-i))) * p**i * (1-p)**(N-i)
    return SUM

print("(N = 10, e=3):",binomialDistribution_greater_than_e(10, 3))
print("(N = 20, e=4):",binomialDistribution_greater_than_e(20, 4))
print("(N = 200, e=40):",binomialDistribution_greater_than_e(200, 40))
print("(N = 1000, e=120):",binomialDistribution_greater_than_e(1000, 120))

```

```python
 #有兩個class
 #所以prior大小是class的數量
prior = np.array([0.35,0.65])   # 初始化先頭的先頭為0，其他為0

#mean是在一種class中 每種feature的絕對值
#所以mean大小是class的數量*feature的數量
mean  = np.array([[1,1],[2,2]])
#cov是在一種class中 每種feature間的差異(covariance matrix)
#所以cov大小是class的數量*feature的數量*feature的數量
cov   = np.array([[[2,1],[1,3]],[[3,-1],[-1,2]]])  # 初始化先頭的先頭為0，其他為0


def discriminant_score(X,class_id):
        x_minus_m = X - mean[class_id]
        #使用課本提供的discriminant_score函式
        discriminant_score = -1.0*0.5 * np.log(np.linalg.det(cov[class_id])) - 0.5 * x_minus_m.T @ np.linalg.inv(cov[class_id]) @ x_minus_m + np.log(prior[class_id])
        return discriminant_score


def predict_proba(X): # 視需要
        #如果輸入是一組data 輸出就會是一組機率
        if len(X.shape) == 1:
            prob_list = []
            for class_id in range(2):
                prob_list.append(np.exp(discriminant_score(X,class_id)))
            return  prob_list/(sum(prob_list)*1.0)

print(predict_proba(np.array([1.5,2.5])))
```
